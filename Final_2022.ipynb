##**ตวจสอบว่ามี NaN หรือ Missing หรือไม่**

```python
data1.isnull().any() 
#.isnull() เพื่อระบุว่านิพจน์มีค่า Null หรือไม่
#.any() เพื่อดูรายละเอียดการเปรียบเทียบข้อมูลให้ออกมาในรูป true false
```

จากผลเราจะพิจารณาจากการอ่านค่า True และ False ถ้า True คือ มีค่า missing ดังนั้น จะเห็นได้ว่าข้อมูลมีค่า missing เกือบในทุกคอลัมน์ ดังนั้น สามารถอนุมานได้เลยว่า ข้อมูลมีค่า missing จำนวนมาก จึงไม่เหมาะกับตัดข้อมูล หรือ .dropna
จึงพิจารณาให้ใช้วิธีแทนค่าเฉพาะคอลัมน์ข้อมูลที่เราต้องการใช้ 

##**ลองใช้ .dropna ซึ่งจะใช้ตัดแถวข้อมูลที่มีค่า missing**

```python
data_drop = data1.dropna() #ใช้ dropna เพื่อต้องการลบข้อมูลทั้งแถวออกเมื่อมีคอลัมน์ใดคอลัมน์หนึ่งเป็น NaN
data_drop
```

```python
removes = data1.shape[0] - data_drop.shape[0]
print(f'size before drop = {data1.shape[0]}')
print(f'size after drop = {data_drop.shape[0]}')
print(f'% from drop = {(removes/data1.shape[0])*100}')
```

ดังนั้น ควรใช้วิธีการแทนค่า แทนการใช้ dropna 

##**ลองใช้การแทนค่าข้อมูลที่มีค่า missing**

```python
ov = data1[['ov']].mean() #กำหนดให้ ov เป็นค่าเฉลี่ยของอายุ
data1[['ov']] = data1[['ov']].fillna(ov) #แทนค่าเฉลี่ยลงในข้อมูลที่ว่าง หรือเป็นค่า missing
```
