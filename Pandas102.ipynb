## ต่อในแนวแกน X

```python
personal_data = data_df[['No.','sex','age','nationality']]
personal_data.head()
```

```python
covid_data = data_df[['No.','announce_date','province_of_isolation','province_of_onset','risk']].sample(frac=1)
covid_data.head()
```

### เอาตารางสองอันมาต่อกัน

```python
personalcovid_data = personal_data.merge(covid_data)     # เอาตาราง personal_data เป็นตารางหลัก แล้วเอาตาราง covid_data มาต่อ
personalcovid_data.head()
```

```python
personalcovid_data = covid_data.merge(personal_data)  # เอาตาราง covid_data เป็นตารางหลัก แล้วเอาตาราง personal_data มาต่อ
personalcovid_data.head()
```

# Save table

# Groupby

https://www.kaggle.com/crawford/python-groupby-tutorial#

```python
data_df.groupby('nationality').count()  # จัด group และนับค่าใน column
```

```python
data_df.groupby('nationality').mean()  # จัด group และคำนวณค่าเฉลี่ยใน column
```

```python
data_df.groupby('nationality')['age'].min()  # min ต้อง เลือก 1 column มาจัด group และอีก 1 column มาเพื่อหาค่า
```

min กับ max เป็น bug ถ้าจะหาค่าต้องใส่ช่องว่างตลอด ' ' แบบนี้

```python
data_df.groupby('nationality').min(' ')  # จัด group และคำนวณค่าน้อยสุดใน column
```

## Dictionary (row oriented)

```python
import pandas as pd
```

```python
data_list = [ {'account':'Jones LLC','Jan':150,'Feb':200,'Mar':140},
                   {'account':'Alpha Co','Jan':200,'Feb':210,'Mar':215},
                   {'account':'Blue Inc','Jan':50,'Feb':90,'Mar':95}
                 ]                                                                  # สร้าง list เพื่อใส่ข้อมูลจะทำตาราง
```

```python
data_list
```

```python
df1 = pd.DataFrame(data_list)
df1
```

## List (row oriented)

```python
data = [ ['Jones LLC',150,200,140],
             ['Alpha Co',200,210,215],
             ['Blue Inc',50,90,95] ]  # ใส่ค่า
header = ['account','Jan','Feb','Mar']          # กำหนดหัว column
```

```python
df2 = pd.DataFrame.from_records(data,columns=header)
df2
```

## Dictionary (column oriented)

```python
data_dict = { 'account':['Jones LLC','Alpha Co','Blue Inc'],
                    'Jan':[150,200,50],
                    'Feb':[200,210,90],
                    'Mar':[140,215,95]
}
```

```python
df3 = pd.DataFrame.from_dict(data_dict)
df3
```
